Split Target or Input Data

```python
x = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x)
```

```python
x_sc = sc.transform(x)
x_sc
```

Train or Test Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, train_size = 0.8, random_state= 42,
                                                   stratify=y)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dt = DecisionTreeClassifier()
```

```python
dt.get_params()
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
def model_evaluation(y_actual, y_prediction):
    print('Accuracy_Score = ',accuracy_score(y_actual, y_prediction))
    print('Confusion Matrix \n',confusion_matrix(y_actual, y_prediction))
    sns.heatmap(confusion_matrix(y_actual, y_prediction), annot = True, fmt = '.2f')
    print('Classification_Report \n',classification_report(y_actual, y_prediction))
```

```python
model_evaluation(y_test, y_pred)
```

```python
dt.get_depth()
```

```python
plt.figure(figsize=(15,10))
plot_tree(dt,filled=True)
plt.savefig('dt')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10]
}
```

```python
grid = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5)
grid.fit(x_train_res, y_train_res)
```

```python
dt = GridSearchCV(dt, param_grid= param_grid, refit=True, verbose = 3)
```

```python
dt.fit(x_train,y_train)
```

```python
print('Best Parameters = ',dt.best_params_)
print('Best Accuracy = ',dt.best_score_)
```

```python
best_dt = DecisionTreeClassifier(criterion='entropy', max_depth = 3)
best_dt.fit(x_train, y_train)
best_dt.score(x_train, y_train)
```

```python
best_y_pred = best_dt.predict(x_test)
```

```python
model_evaluation(y_test, best_y_pred)
```

```python
plt.figure(figsize=(15,10))
plot_tree(best_dt,filled=True)
plt.savefig('best_dt')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(dt, x_train, y_train, cv=5)
print("Cross-validated accuracy:", scores.mean())
```

```python
from sklearn.metrics import roc_auc_score
y_proba = best_dt.predict_proba(x_test)[:, 1]
print("ROC-AUC Score =", roc_auc_score(y_test, y_proba))
```
